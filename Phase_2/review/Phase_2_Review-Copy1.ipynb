Your ability to answer the first question has the executive excited and now she has many other questions about the types of movies being made and the differences in those movies budgets and gross amounts.

Read through the questions below and **determine what type of statistical test you should use** for each question and **write down the null and alternative hypothesis for those tests**.

1. Is there a relationship between the number of Facebook likes for a cast and the box office gross of the movie?
2. Do foreign films perform differently at the box office than non-foreign films?
3. Of all movies created are 40% rated R?
4. Is there a relationship between the language of a film and the content rating (G, PG, PG-13, R) of that film?
5. Is there a relationship between the content rating of a film and its budget? 

```python
# your answers here
#1. One quantitative predictor, quantitavie outcome --> Simple regression
#H0: There is no relationship between # of fb like for a cast and the box office gross.
#HA: There is a relationship between # of fb like for a cast and the box office gross.

#2. Comparing means, population std unknown --> two-sample T-test
#H0: Mean gross of foreign films = mean gross of non-foreign films
#HA: Mean gross of foreign films does not = mean gross of non-foreign films

#3. Calculating proportions, np0>10, n(1-p0)>10 --> Z-test
#H0: The proportion of movies that are R-rated = 40%
#HA: The proportion of movies that are R-rated does not = 40%

#4. Categorial predictor, catorical outcome --> Chi-squared test
#H0: There is not a relationship between the lanugage of a film at the content rating
#HA: There is a relationship between the lanugage of a film at the content rating

#5. More than 2 categorical predictors, 1 quantitative outcome --> ANOVA
#H0: There is not a relationship between the content rating of a film and its budget
#HA: There is  a relationship between the content rating of a film and its budget
```

```python
#test #1:
df_cleaned = df.dropna(axis=0,subset=['gross'])
df_cleaned = df_cleaned.dropna(axis=0,subset=['cast_total_facebook_likes'])


from statsmodels.formula.api import ols

slr_model = ols(formula='gross~cast_total_facebook_likes', data=df_cleaned).fit()
slr_model.summary()
```

```python
#Based on the above, for every 914 likes, a films gross increases $1.
#This relationship accounts for 6.1% of the variation in the data by using total facebook likes
#With 95% confidence, we reject the null hypothesis that there is no relationship between fb likes and gross
```

```python
#test #3:

p_hat = len(df.loc[df['content_rating'] == 'R'])/len(df['content_rating'])
moe = math.sqrt(p_hat*(1-p_hat)/len(df['content_rating']))

lower = p_hat - 1.96 * moe
upper = p_hat + 1.96 * moe

print(lower,upper)
```

```python
#Based on the above, we are 95% sure that the true population proportion of R-rated movies is
#between 40.64% and 43.36%.
#We therefore reject the null hypothesis that the pop proportion of R-rated movies is 40%
#bc 40% is not contained within our 95% confidence interval

```

```python
#test #5:
df_cleaned = df.dropna(axis=0,subset=['content_rating'])
df_cleaned = df_cleaned.dropna(axis=0,subset=['budget'])

stats.f_oneway(df_cleaned['budget'][df_cleaned['content_rating'] == 'R'],
                df_cleaned['budget'][df_cleaned['content_rating'] == 'PG-13'], 
                df_cleaned['budget'][df_cleaned['content_rating'] == 'PG'],
                df_cleaned['budget'][df_cleaned['content_rating'] == 'G'])
```

```python
#Based on the above, we cannot reject the H0 at 95% confidence.
#There is over a 5% chance that there is not a relationship between the content rating of a film and its budget
```

## Question 5

**Bayes Theorem**

An advertising executive is studying television viewing habits of married men and women during prime time hours. Based on the past viewing records he has determined that during prime time wives are watching television 60% of the time. It has also been determined that when the wife is watching television, 40% of the time the husband is also watching. When the wife is not watching the television, 30% of the time the husband is watching the television. Find the probability that if the husband is watching the television, the wife is also watching the television.

```python
# your answer here
#a = wife watching TV
#b = husband watching TV

P_a = 0.6
P_b_given_a = 0.4
P_b_given_notA = 0.3
P_notA = 1 - P_a

P_a_given_b = (P_b_given_a * P_a) / ((P_b_given_a * P_a) + (P_b_given_notA * P_notA))
P_a_given_b
#0.66
```
